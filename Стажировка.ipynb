# **Очистка и работа с данными**

```python
dataset_copy = dataset.copy()
dataset_copy.drop_duplicates(inplace= True)  #Удаление дубликатов
dataset_copy.duplicated().sum()
dataset_copy.shape
dataset_copy.isna().sum()

```

Распаковка вложенных значений ('Процессор', 'Оперативная память', 'Жёсткий диск', 'Габариты товара', 'Габариты товара (с упаковкой)')




```python
Processor_type = dataset_copy['Процессор'][0]
Processor_type1 = Processor_type.replace("'", "").replace("{", "").replace("}", ""). replace(":", "")
Processor_type2 = Processor_type1.split(',')
Processor_type2[1].strip('Количество ядер процессора')

def new_processor(dataset):
  result = dataset.replace("'", "").replace("{", "").replace("}", ""). replace(":", "").split(',')[0].strip('Процессор_тип')
  return result
def processor_core(dataset):
  result = dataset.replace("'", "").replace("{", "").replace("}", ""). replace(":", "").split(',')[1].strip('Количество ядер процессора')
  return result

print(new_processor(dataset_copy['Процессор'][10]))
print(processor_core(dataset_copy['Процессор'][10]))

dataset_copy['Processor'] = dataset_copy['Процессор'].apply(new_processor) # Применение функции ко всему столбцу
dataset_copy['Processor_core'] = dataset_copy['Процессор'].apply(processor_core) # Применение функции ко всему столбцу
dataset_copy.drop(columns = 'Процессор', inplace=True)
dataset_copy.head()
```

Работа с признаком "Оперативная память"

```python
def ram_type(dataset):
  result = dataset.replace("'", "").replace("{", "").replace("}", ""). replace(":", "").split(',')[0].strip('Тип оперативной памяти')
  return result

def volume_ram(dataset):
  result = dataset.replace("'", "").replace("{", "").replace("}", ""). replace(":", "").split(',')[1].strip('Объем оперативной памяти (Гб)')[0:2]
  return result

print(volume_ram(dataset_copy['Оперативная память'][10]))
print(ram_type(dataset_copy['Оперативная память'][10]))
dataset_copy['Ram_type'] = dataset_copy['Оперативная память'].apply(ram_type)
dataset_copy['Volume_ram'] = dataset_copy['Оперативная память'].apply(volume_ram)
result = dataset_copy['Оперативная память'][2].replace("'", "").replace("{", "").replace("}", ""). replace(":", "").split(',')[1].strip('Объем оперативной памяти (Гб)')[0:2]
dataset_copy.drop(columns=['Оперативная память'], inplace=True)
dataset_copy.drop(columns=['Габариты товара'], inplace=True)
dataset_copy.drop(columns=['Габариты товара (с упаковкой)'], inplace=True)
```
